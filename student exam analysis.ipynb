```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
```

```python
print(df.head())
```

```python
df = df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
#data transformation
#df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-oct","5-10")
```

```python
genderdistribution = (df['Gender'] =='female').sum()
print(genderdistribution)
```

```python
plt.figure(figsize = (5,5))
colors = ['red','green']
ax = sns.countplot(data=df,x='Gender',palette=colors,hue = 'Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb = df.groupby("ParentEduc").agg({
    'MathScore':'mean',
    'ReadingScore':'mean',
    'WritingScore':'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.show()
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({
    'MathScore':'mean',
    'ReadingScore':'mean',
    'WritingScore':'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.show()
```

```python

plt.subplot(1,3,1)
sns.boxplot(data = df,x = 'MathScore',color='red')
plt.subplot(1,3,2)
sns.boxplot(data = df,x = 'ReadingScore',color='blue')
plt.tight_layout()
plt.show()
plt.subplot(1,3,3)
sns.boxplot(data = df,x = 'WritingScore',color='orange')
plt.tight_layout()
plt.show()
```

```python
#analysis on ethnic group
print(df['EthnicGroup'].unique())
```

```python
groupa = df.loc[df['EthnicGroup']== "group A"].count()
groupb = df.loc[df['EthnicGroup']== "group B"].count()
groupc = df.loc[df['EthnicGroup']== "group C"].count()
groupd = df.loc[df['EthnicGroup']== "group D"].count()
groupe = df.loc[df['EthnicGroup']== "group E"].count()
l = ['groupa','groupb','groupc','groupd','groupe']
mlist = [groupa['EthnicGroup'],groupb['EthnicGroup'],groupc['EthnicGroup'],groupd['EthnicGroup'],groupe['EthnicGroup']]
print(mlist)
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.show()
```
